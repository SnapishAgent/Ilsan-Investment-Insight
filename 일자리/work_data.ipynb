```python
import pandas as pd
import glob
import os

# CSV 파일이 있는 디렉토리 경로 정의
directory_path = r'C:\Users\twoimo\Documents\GitHub\NewTownInsight\일자리'

# glob을 사용하여 디렉토리 내의 모든 CSV 파일 찾기
csv_files = glob.glob(os.path.join(directory_path, '*.csv'))\

# 모든 CSV 파일을 읽어와서 하나의 DataFrame으로 합치기
combined_df = pd.concat([pd.read_csv(file, encoding='EUC-KR') for file in csv_files], axis=1, ignore_index=False)

# Unnamed: 4 컬럼명 모두 제거
combined_df = combined_df.loc[:, ~combined_df.columns.str.contains('^Unnamed')]

display(combined_df.head())
```

```python
combined_df = combined_df.drop(['분류별', '단위'], axis=1, inplace=False)
combined_df
```
